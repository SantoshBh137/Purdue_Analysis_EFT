```python
def plot_observables(obs, obs_sm_file, xlabel, title, weights_dict, num_bins=20):
    fig, (ax, ax_ratio) = plt.subplots(2, 1, figsize=(12, 8),
                                       gridspec_kw={'height_ratios': [3, 1]}, sharex=True)

    # Determine bin edges from full observable range
    min_val = min(ak.min(obs), ak.min(obs_sm_file))
    max_val = max(ak.max(obs), ak.max(obs_sm_file))
    bin_edges = np.linspace(min_val, max_val, num_bins + 1)
    bin_widths = np.diff(bin_edges)
    bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

    # Color cycle
    colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

    # Extract weights
    weights_reweighted = weights_dict['reweighted_SM']

    # Histograms
    counts_reweighted, _ = np.histogram(obs, bins=bin_edges, weights=weights_reweighted)
    counts_sm_file, _ = np.histogram(obs_sm_file, bins=bin_edges)

    # Normalize
    norm_reweighted = np.sum(counts_reweighted)
    norm_sm_file = np.sum(counts_sm_file)
    stat_reweighted = np.sqrt(counts_reweighted) / norm_reweighted / bin_widths
    stat_sm_file = np.sqrt(counts_sm_file) / norm_sm_file / bin_widths

    # Plot histograms
    count_reweighted,_,_=ax.hist(obs, bins=bin_edges, weights=weights_reweighted, histtype='step',
            label="Reweighted_SM (Madgraph@LO)", color=colors[0], density=True)

    count_sm,_,_=ax.hist(obs_sm_file, bins=bin_edges, histtype='step', linestyle='dashed',
            label="SM (Powheg@NLO)", color='red', density=True)

   

    # Ratio and total uncertainty
    ratio = np.divide(count_reweighted, count_sm, out=np.ones_like(count_sm),
                      where=count_sm!= 0)

    total_uncertainty = np.sqrt(stat_reweighted**2 + stat_sm_file**2)

    # Step-style ratio plot
    extended_ratio = np.append(ratio, ratio[-1])
    extended_uncertainty_up = np.append(ratio + total_uncertainty, ratio[-1] + total_uncertainty[-1])
    extended_uncertainty_down = np.append(ratio - total_uncertainty, ratio[-1] - total_uncertainty[-1])

    ax_ratio.step(bin_edges, extended_ratio, where='post', color='red', label="LO/NLO")
    ax_ratio.fill_between(bin_edges, extended_uncertainty_down, extended_uncertainty_up,
                          step='post', facecolor='none', edgecolor='black',hatch='//////////',linewidth=0.0)

    # Cosmetics
    ax_ratio.set_ylim(0.95, 1.05)
    ax.set_ylabel("Normalized Events")
    ax.set_title(title)
    ax.legend(loc='best')
    plt.style.use(hep.style.CMS)
    ax_ratio.set_xlabel(xlabel)
    ax_ratio.axhline(1, color='gray', linestyle='--', linewidth=1)
    ax_ratio.set_ylabel("LO / NLO")

    plt.tight_layout()
    plt.show()
```
