```python

def extract_observables_from_files(file_sm, file_eft):
    # Define the mask based on the gen_ttbar_mass observable

    # Extracting and applying the mask to observables from the SM file
    observables_sm = {
        'gen_top_pt': file_sm['gen_top_pt'].array(),
        'gen_top_phi': file_sm['gen_top_phi'].array(),
        'gen_top_eta': file_sm['gen_top_eta'].array(),
        'gen_top_mass': file_sm['gen_top_mass'].array(),
        'gen_tbar_pt': file_sm['gen_tbar_pt'].array(),
        'gen_tbar_phi': file_sm['gen_tbar_phi'].array(),
        'gen_tbar_eta': file_sm['gen_tbar_eta'].array(),
        'gen_tbar_mass': file_sm['gen_tbar_mass'].array(),
        'gen_l_pt': file_sm['gen_l_pt'].array(),
        'gen_l_eta': file_sm['gen_l_eta'].array(),
        'gen_l_phi': file_sm['gen_l_phi'].array(),
        'gen_l_mass': file_sm['gen_l_mass'].array(),
        'gen_lbar_pt': file_sm['gen_lbar_pt'].array(),
        'gen_lbar_eta': file_sm['gen_lbar_eta'].array(),
        'gen_lbar_phi': file_sm['gen_lbar_phi'].array(),
        'gen_lbar_mass': file_sm['gen_lbar_mass'].array(),
        'gen_llbar_delta_phi': file_sm['gen_llbar_delta_phi'].array(),
        'gen_llbar_delta_eta': file_sm['gen_llbar_delta_eta'].array(),
        'gen_b1k': file_sm['gen_b1k'].array(),
        'gen_b2k': file_sm['gen_b2k'].array(),
        'gen_b1r': file_sm['gen_b1r'].array(),
        'gen_b2r': file_sm['gen_b2r'].array(),
        'gen_b1n': file_sm['gen_b1n'].array(),
        'gen_b2n': file_sm['gen_b2n'].array(),
        'gen_c_kk': file_sm['gen_c_kk'].array(),
        'gen_c_rr': file_sm['gen_c_rr'].array(),
        'gen_c_nn': file_sm['gen_c_nn'].array(),
        'gen_c_rk': file_sm['gen_c_rk'].array(),
        'gen_c_kr': file_sm['gen_c_kr'].array(),
        'gen_c_nr': file_sm['gen_c_nr'].array(),
        'gen_c_rn': file_sm['gen_c_rn'].array(),
        'gen_c_nk': file_sm['gen_c_nk'].array(),
        'gen_c_kn': file_sm['gen_c_kn'].array(),
        'gen_ll_cHel': file_sm['gen_ll_cHel'].array(),
    }

    # Extracting and applying the mask to observables from the EFT file
    observables_eft = {
        'gen_top_pt': file_eft['gen_top_pt'].array(),
        'gen_top_phi': file_eft['gen_top_phi'].array(),
        'gen_top_eta': file_eft['gen_top_eta'].array(),
        'gen_top_mass': file_eft['gen_top_mass'].array(),
        'gen_tbar_pt': file_eft['gen_tbar_pt'].array(),
        'gen_tbar_phi': file_eft['gen_tbar_phi'].array(),
        'gen_tbar_eta': file_eft['gen_tbar_eta'].array(),
        'gen_tbar_mass': file_eft['gen_tbar_mass'].array(),
        'gen_l_pt': file_eft['gen_l_pt'].array(),
        'gen_l_eta': file_eft['gen_l_eta'].array(),
        'gen_l_phi': file_eft['gen_l_phi'].array(),
        'gen_l_mass': file_eft['gen_l_mass'].array(),
        'gen_lbar_pt': file_eft['gen_lbar_pt'].array(),
        'gen_lbar_eta': file_eft['gen_lbar_eta'].array(),
        'gen_lbar_phi': file_eft['gen_lbar_phi'].array(),
        'gen_lbar_mass': file_eft['gen_lbar_mass'].array(),
        'gen_llbar_delta_phi': file_eft['gen_llbar_delta_phi'].array(),
        'gen_llbar_delta_eta': file_eft['gen_llbar_delta_eta'].array(),
        'gen_b1k': file_eft['gen_b1k'].array(),
        'gen_b2k': file_eft['gen_b2k'].array(),
        'gen_b1r': file_eft['gen_b1r'].array(),
        'gen_b2r': file_eft['gen_b2r'].array(),
        'gen_b1n': file_eft['gen_b1n'].array(),
        'gen_b2n': file_eft['gen_b2n'].array(),
        'gen_c_kk': file_eft['gen_c_kk'].array(),
        'gen_c_rr': file_eft['gen_c_rr'].array(),
        'gen_c_nn': file_eft['gen_c_nn'].array(),
        'gen_c_rk': file_eft['gen_c_rk'].array(),
        'gen_c_kr': file_eft['gen_c_kr'].array(),
        'gen_c_nr': file_eft['gen_c_nr'].array(),
        'gen_c_rn': file_eft['gen_c_rn'].array(),
        'gen_c_nk': file_eft['gen_c_nk'].array(),
        'gen_c_kn': file_eft['gen_c_kn'].array(),
        'gen_ll_cHel': file_eft['gen_ll_cHel'].array(),
    }

    return observables_sm, observables_eft

def plot_observables(obs1, obs2, xlabel, title, weight_sm, weight_eft, num_bins=20,):
    fig, (ax, ax_ratio) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]}, sharex=True)

    # Determine common bin edges based on the min and max values of both observables
    min_val = min(ak.min(obs1), ak.min(obs2))
    max_val = max(ak.max(obs1), ak.max(obs2))
    bin_edges = np.linspace(min_val, max_val, num_bins + 1)

    # Plot the first set of observables (SM) with corresponding weights
    counts1, bins1, _ = ax.hist(obs1, bins=bin_edges, 
                                label='gen_sm', color='blue', weights=weight_sm, histtype='step', density=True)
    
    # Plot the second set of observables (EFT) with corresponding weights
    counts2, bins2, _ = ax.hist(obs2, bins=bin_edges,
                                label='gen_EFT', color='red', weights=weight_eft, histtype='step', density=True)

    # Calculate the ratio between SM and weighted EFT
    bin_centers = 0.5 * (bins1[:-1] + bins1[1:])
    ratio = np.divide(counts1, counts2, out=np.ones_like(counts1), where=counts2 != 0)

    # Plot the ratio for weighted EFT
    ax_ratio.plot(bin_centers, ratio, color='black', marker='o', linestyle='none', label='SM/EFT (weighted)')
    ax_ratio.set_ylabel('SM/EFT Ratio')
    ax_ratio.set_ylim(0.8, 1.2)  # Adjust the limits as needed

    # Adding labels, title, and legend
    ax.set_ylabel("Weighted events (normalized)")
    ax.set_title(title)
    ax.legend(loc='best')

    # Add CMS label
    hep.cms.label(ax=ax)

    # Label for x-axis on the ratio plot
    ax_ratio.set_xlabel(xlabel)
    ax_ratio.axhline(1, color='gray', linestyle='--', linewidth=1)
    ax_ratio.legend(loc='best')  # Add legend for the ratio plot
    plt.tight_layout()
    plt.show()

# Extract observables
observables_sm, observables_eft = extract_observables_from_files(file_eft, file_eft)

observables_to_plot = [
    ('gen_top_pt', 'Top $p_T$ (GeV)', 'Top $p_T$ Distribution'),
    ('gen_top_phi', 'Top $\phi$', 'Top $\phi$ Distribution'),
    ('gen_top_eta', 'Top $\eta$', 'Top $\eta$ Distribution'),
    ('gen_top_mass', 'Top Mass (GeV)', 'Top Mass Distribution'),
    ('gen_tbar_pt', 'Anti-Top $p_T$ (GeV)', 'Anti-Top $p_T$ Distribution'),
    ('gen_tbar_phi', 'Anti-Top $\phi$', 'Anti-Top $\phi$ Distribution'),
    ('gen_tbar_eta', 'Anti-Top $\eta$', 'Anti-Top $\eta$ Distribution'),
    ('gen_tbar_mass', 'Anti-Top Mass (GeV)', 'Anti-Top Mass Distribution'),
    ('gen_tbar_mass', 'Anti-Top Mass (GeV)', 'Anti-Top Mass Distribution'),
    ('gen_l_pt', 'Lepton $p_T$ (GeV)', 'Lepton $p_T$ Distribution'),
    ('gen_l_eta', 'Lepton $\eta$', 'Lepton $\eta$ Distribution'),
    ('gen_l_phi', 'Lepton $\phi$', 'Lepton $\phi$ Distribution'),
    ('gen_l_mass', 'Lepton Mass (GeV)', 'Lepton Mass Distribution'),
    ('gen_lbar_pt', 'Anti-Lepton $p_T$ (GeV)', 'Anti-Lepton $p_T$ Distribution'),
    ('gen_lbar_eta', 'Anti-Lepton $\eta$', 'Anti-Lepton $\eta$ Distribution'),
    ('gen_lbar_phi', 'Anti-Lepton $\phi$', 'Anti-Lepton $\phi$ Distribution'),
    ('gen_lbar_mass', 'Anti-Lepton Mass (GeV)', 'Anti-Lepton Mass Distribution'),
    ('gen_llbar_delta_phi', '$\Delta \phi_{ll}$', '$\Delta \phi_{ll}$ Distribution'),
    ('gen_llbar_delta_eta', '$\Delta \eta_{ll}$', '$\Delta \eta_{ll}$ Distribution'),
    ('gen_b1k', 'gen_b1k', 'gen_b1k'),
    ('gen_b2k', 'gen_b2k', 'gen_b2k'),
    ('gen_b1r', 'gen_b1r', 'gen_b1r'),
    ('gen_b2r', 'gen_b2r', 'gen_b2r'),
    ('gen_b1n', 'gen_b1n', 'gen_b1n'),
    ('gen_b2n', 'gen_b2n', 'gen_b2n'),
    ('gen_c_kk', 'gen_c_kk', 'gen_c_kk'),
    ('gen_c_rr', 'gen_c_rr', 'gen_c_rr'),
    ('gen_c_nn', 'gen_c_nn', 'gen_c_nn'),
    ('gen_c_rk', 'gen_c_rk', 'gen_c_rk'),
    ('gen_c_kr', 'gen_c_kr', 'gen_c_kr'),
    ('gen_c_nr', 'gen_c_nr', 'gen_c_nr'),
    ('gen_c_rn', 'gen_c_rn', 'gen_c_rn'),
    ('gen_c_nk', 'gen_c_nk', 'gen_c_nk'),
    ('gen_c_kn', 'gen_c_kn', 'gen_c_kn'),
    ('gen_ll_cHel', 'gen_ll_cHel', 'gen_ll_cHel')
]

# Loop through each observable and plot
for obs_key, xlabel, title in observables_to_plot:
    plot_observables(observables_sm[obs_key],observables_eft[obs_key], xlabel, title, weight_sm, weight_ctGRe1)
```
