```python
import numpy as np
import matplotlib.pyplot as plt

# Define the integrand
def integrand(x, y):
    return np.exp(-(x**2 + y**2))

# Monte Carlo integration function
def monte_carlo_integration(func, x_range, y_range, num_samples):
    x_min, x_max = x_range
    y_min, y_max = y_range
    
    x_random = np.random.uniform(x_min, x_max, num_samples)
    y_random = np.random.uniform(y_min, y_max, num_samples)
    
    function_values = func(x_random, y_random)
    area = (x_max - x_min) * (y_max - y_min)
    
    integral = area * np.mean(function_values)
    return integral

# Analytical value of the integral
analytical_value = 0.557746

# Parameters
x_range = (0, 1)
y_range = (0, 1)
sample_sizes = np.logspace(2, 9, num=20, dtype=int)  # Logarithmically spaced sample sizes

# Store results for plotting
estimated_values = []

for num_samples in sample_sizes:
    estimated_value = monte_carlo_integration(integrand, x_range, y_range, num_samples)
    estimated_values.append(estimated_value)

# Plot the convergence
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, estimated_values, label="Monte Carlo Estimate", marker="o")
plt.axhline(y=analytical_value, color="r", linestyle="--", label="Analytical Value")
plt.xscale("log")
plt.xlabel("Number of Samples (log scale)", fontsize=12)
plt.ylabel("Integral Value", fontsize=12)
plt.title("Convergence of Monte Carlo Integration", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
